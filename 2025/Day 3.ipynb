```python
import re

with open('3-1.txt', 'r') as f:
    data = f.read()
```

```python
def multiply_matches(m, _text):
    _op = _text[m.span()[0]:m.span()[1]]
    numbers = _op.replace('mul(', '').replace(')', '').split(',')
    return int(numbers[0]) * int(numbers[1])
```

```python
matches = re.finditer('mul\(\d{1,3},\d{1,3}\)', data)
print(sum([multiply_matches(match, data) for match in matches]))
```

```python
matches = re.finditer('mul\(\d{1,3},\d{1,3}\)|don\'t\(\)|do\(\)', data)
```

```python
enabled = True
sums = []
for i, match in enumerate(matches):
    _op = data[match.span()[0]:match.span()[1]]
    if _op == 'do()':
        enabled = True
    elif _op == "don't()":
        enabled = False
    else:
        if enabled:
            sums.append(multiply_matches(match, data))
            
print(sum(sums))
            
```
