```python
import re

with open('6-1.txt', 'r') as f:
    data = f.read()
```

```python
grid = data.replace('\n\n', '\n').strip().split('\n')
```

```python
grid = [
    '....#.....',
    '.........#',
    '..........',
    '..#.......',
    '.......#..',
    '..........',
    '.#..^.....',
    '........#.',
    '#.........',
    '......#...'
]
```

```python
def move_guard(direction, pos_x, pos_y, _grid):
    if direction == '^': # try and move up (x - 1)
        if pos_x - 1 < 0:
            return False
        elif _grid[pos_x - 1][pos_y] != '#':
            return direction, pos_x - 1, pos_y
        else:
            return '>', pos_x, pos_y
    if direction == '>': # try and move right (y + 1)
        if pos_y + 1 >= len(_grid[pos_x]):
            return False
        elif _grid[pos_x][pos_y + 1] != '#':
            return direction, pos_x, pos_y + 1
        else:
            return 'v', pos_x, pos_y
    if direction == 'v': # try and move down (x + 1)
        if pos_x + 1 >= len(_grid):
            return False
        elif _grid[pos_x + 1][pos_y] != '#':
            return direction, pos_x + 1, pos_y
        else:
            return '<', pos_x, pos_y
    if direction == '<': # try and move left (y - 1)
        if pos_y - 1 < 0:
            return False
        elif _grid[pos_x][pos_y - 1] != '#':
            return direction, pos_x, pos_y - 1
        else:
            return '^', pos_x, pos_y
        
    
def find_start(_grid):
    for x, row in enumerate(_grid):
        for y, cell in enumerate(row):
            if cell == '^':
                return x, y

```

```python
is_still_searching = True
directions = set()
direction, pos_x, pos_y = "^", *find_start(grid)
directions.add((pos_x, pos_y))
while is_still_searching:
    result = move_guard(direction, pos_x, pos_y, _grid=grid)
    if not result:
        is_still_searching = False
        print(f'leaving after moving in direction: {direction} from {pos_x}, {pos_y}')
        break
#     print(f'moving in direction: {direction} to {pos_x}, {pos_y}')        
    direction, pos_x, pos_y = result
    directions.add((pos_x, pos_y))

print(f'there are {len(directions)} unique positions')
```

```python
from copy import deepcopy
is_still_searching = True
direction, pos_x, pos_y = "^", *find_start(grid)
start_x, start_y = pos_x, pos_y
loop_count = 0
        
obstacle_positions = directions - {(start_x - 1, start_y)}
# obstacle_positions = [(3,4), (6,3), (7,6)]  

for obs_x, obs_y in obstacle_positions:
    print(f'testing obstacle @ {obs_x}, {obs_y}')
    visited_states = {('^', start_x, start_y)}
    grid_copy = deepcopy(grid)
    grid_copy[obs_x] = grid_copy[obs_x][:obs_y] + '#' + grid_copy[obs_x][obs_y + 1:]
    direction, pos_x, pos_y = '^', start_x, start_y
    is_still_searching = True
    loop_found = False
    while is_still_searching and not loop_found:
        result = move_guard(direction, pos_x, pos_y, grid_copy)
        if not result:
#             print(f'{result} - leaving after moving in direction: {direction} from {pos_x}, {pos_y}')
            is_still_searching = False
            loop_found = False
        else:       
            direction, pos_x, pos_y = result
            if (direction, pos_x, pos_y) in visited_states:
                print(f'loop found with obstacle: {obs_x}, {obs_y}')
                loop_count += 1
                loop_found = True
            visited_states.add((direction, pos_x, pos_y))

print(f'obstacle  creating loop count: {loop_count}')
```
